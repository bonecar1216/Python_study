## Dataset Description
See, fork, and run a random forest benchmark model through Kaggle Scripts

You are provided hourly rental data spanning two years. For this competition, the training set is comprised of the first 19 days of each month, while the test set is the 20th to the end of the month. You must predict the total count of bikes rented during each hour covered by the test set, using only information available prior to the rental period.

## Data Fields
* datetime - hourly date + timestamp  
* season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
* holiday - whether the day is considered a holiday
* workingday - whether the day is neither a weekend nor holiday
* weather - 
    * 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    * 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    * 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    * 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
* temp - temperature in Celsius
* atemp - "feels like" temperature in Celsius
* humidity - relative humidity
* windspeed - wind speed
* casual - number of non-registered user rentals initiated
* registered - number of registered user rentals initiated
* count - number of total rentals

## Evaluation
Submissions are evaluated one the Root Mean Squared Logarithmic Error (RMSLE)

```python
# 연도와 월을 붙이는 함수
def concatenate_year_month(datetime):
    return '{0}-{1}'.format(datetime.year, datetime.month)

train['year_month'] = train['datetime'].apply(concatenate_year_month)

print(train.shape)
train[['datetime', 'year_month']].head()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
fig.set_size_inches(18, 4)

sns.barplot(data = train, x = 'year', y = 'count', ax = ax1)
sns.barplot(data = train, x = 'month', y = 'count', ax = ax2)

fig, ax3 = plt.subplots(nrows = 1, ncols = 1)
fig.set_size_inches(18,4)

sns.barplot(data = train, x = 'year_month', y = 'count', ax = ax3)
```

* 2011년도보다 2012년도 자전거 대여량이 많음.
* 여름~가을의 대여량이 많으며, 겨울이 되면서 대여량이 감소하는 경향을 보임.
* 시간이 지나면서 전제적으로 자전거 대여량이 증가하는 추세임.

```python
#Outlier 제거
trainWithoutOutliers = train[np.abs(train['count'] - train['count'].mean()) <= (3*train['count'].std())]

print(train.shape)
print(trainWithoutOutliers.shape)
```

* 자전거 대여량 데이터에서 0에 몰려있는 값들과 너무 큰 값들을 제거
    * 총 10,886개의 데이터에서 147개의 데이터가 제거되어 10,739개의 데이터가 남았음.
    * 전체 데이터의 1.4%정도가 제거되었으며, 제거된 데이터의 양이 적은 것은 아님.
    * 해당 데이터를 제거하는 것이 예측의 정확도를 높히는지, 아니면 다른 방법으로 데이터를 처리하는 것이 좋은지 비교해보는 것이 필요해 보임.

```python
# Count값의 데이터 분포도를 파악

figure, axes = plt.subplots(ncols = 2, nrows = 2)
figure.set_size_inches(12, 10)

sns.distplot(train['count'], ax = axes[0][0])
stats.probplot(train['count'], dist = 'norm', fit = True, plot = axes[0][1])
sns.distplot(np.log(trainWithoutOutliers['count']), ax = axes[1][0])
stats.probplot(np.log(trainWithoutOutliers['count']), dist = 'norm', fit = True, plot = axes[1][1])
```

* outlier처리 이전에는 자전거 대여량의 데이터가 왼쪽으로 많이 치우쳐 있는 형상을 하고 있음. outlier처리 이후에는 정규분포를 갖지는 않지만, 처리 이전보다는 분포가 정규 분포에 근접할 것을 확인할 수 있음.
    * 대부분의 머신러닝에서는 종속변수가 정규분포를 갖는 것이 바람직함.
* QQplot를 보면 outlier처리 이전에는 왼쪽으로 많이 치우친 경향을 보였지만, outlier 처리 이후에는 정규 분포에 조금 가까워졌지만 오른쪽으로 치우친 경향을 갖고 있음을 확인함.

* 측정이 되지 않은 데이터들이 0으로 입력된 것으로 예상됨.
    * 측정이 되지 않은 데이터들에 평균값을 넣어주는 방법이 있으나, 0으로 되어있는 값이 많기 때문에 평균으로 바꿔주는 것은 예측 데이터가 부정확해질 수 있을 것 같음.

```python
# windspeed가 0이 아닌 것의 세트를 나누어 준다.
trainWind0 = train.loc[train['windspeed'] == 0]
trainWindNot0 = train.loc[train['windspeed'] !=0]
print(trainWind0.shape)
print(trainWindNot0.shape)
```

```python
# 풍속을 머신러닝을 통해 예측하여 값을 입력
from sklearn.ensemble import RandomForestClassifier

def predict_windspeed(data):
    
    # 풍속이 0인것과 아닌 것을 나누어 준다.
    dataWind0 = data.loc[data['windspeed']==0]
    dataWindNot0 = data.loc[data['windspeed']!=0]
    
    # 풍속을 예측할 feature를 선택한다.
    wCol = ['season', 'weather', 'humidity', 'month', 'temp', 'year', 'atemp']
    
    # 풍속이 0이 아닌 데이터들의 타입을 str으로 바꿔준다.
    dataWindNot0['windspeed'] = dataWindNot0['windspeed'].astype('str')
    
    # 랜덤포레스트 분류기를 사용한다.
    rfModel_wind = RandomForestClassifier()
    
    # wCol에 있는 feature를 이용해서 분류기를 학습시킨다.
    rfModel_wind.fit(dataWindNot0[wCol], dataWindNot0['windspeed'])
    
    # 학습한 값을 바탕으로 풍속이 0으로 기록된 데이터의 풍속을 예측한다.
    wind0values = rfModel_wind.predict(X = dataWind0[wCol])
    
    # 값을 다 예측한 후 비교
    predictWind0 = dataWind0
    predictWindNot0 = dataWindNot0
    
    # 값이 0으로 기록된 풍속에 대해 예측한 값을 넣어준다.
    predictWind0['windspeed'] = wind0values
    
    # dataWindNot0에 0이 아닌 풍속이 있는 데이터프레임에 예측한 값이 있는 데이터 프레임을 합쳐준다.
    data = predictWindNot0.append(predictWind0)
    
    # 풍속의 데이터타입을 float으로 바꿔준다.
    data['windspeed'] = data['windspeed'].astype('float')
    
    data.reset_index(inplace = True)
    data.drop('index', inplace = True, axis = 1)
    
    return data
```

```python
# 0값을 조정한다.
train = predict_windspeed(train)
# test 데이터의 windspeed 0값을 예측하여 입력한 경우 점수가 낮게 나와서 제외

fig, ax1 = plt.subplots()
fig.set_size_inches(18,6)

plt.sca(ax1)
plt.xticks(rotation = 30, ha = 'right')
ax1.set(ylabel = 'Count', title = 'train_windspeed')
sns.countplot(data = train, x = 'windspeed', ax = ax1)
```

## Score

## RMSLE

```python
from sklearn.metrics import make_scorer

def rmsle(predicted_values, actual_values):
    # 넘파이로 배열 형태 변환
    predicted_values = np.array(predicted_values)
    actual_values = np.array(actual_values)
    
    # 예측값과 실제값에 1을 더하고 로그를 씌워준다.
    log_predicted = np.log(predicted_values + 1)
    log_actual = np.log(actual_values + 1)
    
    # 위에서 계산한 예측값에서 실제값을 빼주고 제곱한다.
    diff = (log_predicted - log_actual)**2
    
    # 차이값의 제곱을 평균한 값에 루트를 씌운다.
    score = np.sqrt(diff.mean())
    
    return score

rmsle_scorer = make_scorer(rmsle)
rmsle_scorer
```

```python
# K-Fold방법으로 교차 검증 진행
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

k_fold = KFold(n_splits = 10, shuffle = True, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor

max_depth_list = []

model = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=0)

model
```

```python
%time score = cross_val_score(model, X_train, y_train, cv = k_fold, scoring = rmsle_scorer)
score = score.mean()
# 0에 근접할수록 좋은 데이터
print('Score = {0:.5f}'.format(score))
```
