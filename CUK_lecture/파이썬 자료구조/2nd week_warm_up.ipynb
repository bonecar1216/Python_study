# 반복문

2. range() 구문을 이용하여 sample_list를 아래와 같이 출력하라.

sample_list = ["가", "나", "다", "라"]

# 함수

2개의 숫자를 파라미터로 전달받아
전달 받은 숫자를 덧셈(addition), 뺄샘(subtraction), 곱셈(multiplication), 나눗셈(division)하여 리턴하는 함수를 각각 작성하라.

단, 뺄셈, 나눗셈은 전달 인자의 순서대로 수행할 것,
예) subtraction(a,b)   ---> a-b
예) division(a,b)      ---> a/b    

````python
# 문제 답안
def addition(a,b):
    return a+b

def subtraction(a,b):
    return a-b

def multiplication(a,b):
    return a*b

def division(a,b):
    return a/b
````

하나의 리스트를 입력받아 홀수만 화면에 출력하는 print_even 함수를 정의하라.



```
print_even ([1, 3, 2, 10, 12, 11, 15])
1
3
11
15
```



````python
# 문제 답안

def print_even(list):
    for a in list:
        if a%2 == 1:
            print(a)

print_even ([1, 3, 2, 10, 12, 11, 15])
````

# 클래스

위에서 작성한 사칙 연산 함수를 메서드로 사용하는, Calc 클래스를 작성하라. 

단, 숫자 2개를 입력으로 받아 초기화하는 생성자를 포함하라.

````python
# 문제 답안
class Calc():
    
    def __init__(self, a, b):
        self.a = a
        self.b = b
        print(self.a, self.b)

    def addition(self):
        return self.a+self.b

    def subtraction(self):
        return self.a-self.b

    def multiplication(self):
        return self.a*self.b

    def division(self):
        return self.a/self.b

````

````python
a = Calc(1,2)
print(a.addition())
print(a.subtraction())
````
