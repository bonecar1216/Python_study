### 2.4 Age

```python
print('제일 나이 많은 탑승객: {:.1f} Years'.format(df_train['Age'].max()))
print('제일 어린 탑승객: {:.1f} Years'.format(df_train['Age'].min()))
print('탑승객 평균 나이: {:.1f} Years'.format(df_train['Age'].mean()))
```

```python
fig, ax = plt.subplots(1,1, figsize = (9,5))
sns.kdeplot(df_train[df_train['Survived']==1]['Age'], ax = ax)
sns.kdeplot(df_train[df_train['Survived']==0]['Age'], ax = ax)
plt.legend(['Survived == 1', 'Survived == 0'])
plt.show()
```

* 생존자 중 나이 어린 경우가 많음.

```python
plt.subplots(figsize = (10,6))
df_train['Age'][df_train['Pclass'] == 1].plot(kind = 'kde')
df_train['Age'][df_train['Pclass'] == 2].plot(kind = 'kde')
df_train['Age'][df_train['Pclass'] == 3].plot(kind = 'kde')

plt.xlabel('Age')
plt.ylabel('Age Distribution withiin Classes')
plt.legend(['1st Class', '2nd Class', '3rd Class'])
```

* Class가 높을수록 나이가 많은 사람의 비율이 증가함.

```python
cumulate_survival_ratio = []
for i in range(1, 80):
    cumulate_survival_ratio.append(df_train[df_train['Age']<i]['Survived'].sum()/len(df_train[df_train['Age']<i]['Survived']))

plt.figure(figsize = (7,7))
plt.plot(cumulate_survival_ratio)
plt.title('Survival rate change depending on range of Age', y=1.02)
plt.ylabel('Survival rate')
plt.xlabel('Range of Age(0~x)')
plt.xticks(range(0,100,20))
plt.show()
```

* 나이가 어릴수록 생존률이 높음 -> feature로 사용

### 2.7 Family - SibSp (형제, 자매) + Parch (부모, 자녀)

* SibSp와 Parch를 합하여 Family로 만든 후 분석 진행

```python
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch'] + 1 # 자신을 포함
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch'] + 1 # 자신을 포함
```

```python
print('Maximum size of Familiy: ', df_train['FamilySize'].max())
print('Minumum size of Familiy: ', df_train['FamilySize'].min())
```

```python
f, ax = plt.subplots(1,3, figsize = (40,10))
sns.countplot(x='FamilySize', data = df_train, ax = ax[0])
ax[0].set_title('(1) No. of Passengers Boarded', y = 1.02)

sns.countplot(x='FamilySize', hue = 'Survived', data = df_train, ax = ax[1])
ax[1].set_title('(2) Survived countplot depending on FamilySize', y = 1.02)

df_train[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index = True).mean().sort_values(by='Survived', ascending = False).plot.bar(ax = ax[2])
ax[2].set_title('(3) Survived rate depending on FamilySize', y = 1.02)

plt.subplots_adjust(wspace = 0.2, hspace = 0.5)
plt.show()
```

* Figure(1): 가족 크기가 1~11까지 있음. 대부분 1명이며, 2,3,4명의 순임.
* Figure(2), (3): 가족의 크기가 4인 경우, 가장 생존률이 높으며, 2~4명 정도의 가족 크기를 갖는 경우 생존 확률이 높음. 가족 크기가 1인 경우나 너무 큰 경우 (5명 이상)는 생존 확률이 낮아짐.

### 2.8 Fare

```python
fig, ax = plt.subplots(1,1, figsize = (8,8))
g = sns.distplot(df_train['Fare'], color ='b', label = 'Skewness: {:.2f}'.format(df_train['Fare'].skew(), ax = ax))
g = g.legend(loc = 'best')
```

* 매우 비대칭인 분포를 갖고 있음. 이러한 데이터를 모델에 넣어줄 경우, 모델 학습이 잘 안될 수도 있음. (이상치에 민감하게 반응)
* 이상치의 영향을 줄이기 위해 Fare값에 log를 취함.

```python
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean() # 테스트 데이터의 Fare에 있는 null값을 평균값으로 치환

df_train['Fare'] = df_train['Fare'].map(lambda i:np.log(i) if i > 0 else 0)
df_test['Fare'] = df_test['Fare'].map(lambda i:np.log(i) if i > 0 else 0)
```

```python
fig, ax = plt.subplots(1,1, figsize = (8,8))
g = sns.distplot(df_train['Fare'], color ='b', label = 'Skewness: {:.2f}'.format(df_train['Fare'].skew(), ax = ax))
g = g.legend(loc = 'best')
```

* 로그를 취한 후, 비대칭성이 많이 사라진 것을 확인.
