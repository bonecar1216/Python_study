https://www.acmicpc.net/problem/4949

```python
while True:
    list_p = []
    sentence = input()
    if sentence == '.':
        break

    ans = True
    for s in sentence:
        if s == '(' or s == '[':
            list_p.append(s)
        elif s == ')':
            if len(list_p) > 0 and list_p[-1] == '(':
                list_p.pop(-1)
            else:
                ans = False
                continue
        elif s == ']':
            if len(list_p) > 0 and list_p[-1] == '[':
                list_p.pop(-1)
            else:
                ans = False
                continue
    if ans == False or len(list_p) != 0:
        print('no')
    else:
        print('yes')
```
