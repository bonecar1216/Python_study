https://www.acmicpc.net/problem/28278

```python
def stack(a, b = None):
    global stack_list
    if a == 1:
        stack_list.append(b)
    elif a == 2:
        if len(stack_list) != 0:
            return(stack_list.pop())
        else:
            return(-1)
    elif a == 3:
        return len(stack_list)
    elif a == 4:
        if len(stack_list) != 0:
            return 0
        else:
            return 1
    elif a == 5:
        if len(stack_list) != 0:
            return stack_list[-1]
        else:
            return -1


N = int(input())
stack_list = []
for _ in range(N):
    str_num = input()
    if ' ' in str_num :
        a, b = map(int,str_num.split(' '))
    else:
        a = int(str_num)
    ans = stack(a, b)
    if a != 1:
        print(ans)
```

* 다시 시간 초과 발생
* input문을 sys.stdin.readline().split() 사용하여 수정하였고, 제출 완료

```python
import sys

N = int(input())
stack_list = []
for _ in range(N):
    num = sys.stdin.readline().split()
    if num[0] == '1':
        stack_list.append(int(num[1]))
    elif num[0] == '2':
        if len(stack_list) != 0:
            print(stack_list.pop())
        else:
            print(-1)
    elif num[0] == '3':
        print(len(stack_list))
    elif num[0] == '4':
        if len(stack_list) != 0:
            print(0)
        else:
            print(1)
    elif num[0] == '5':
        if len(stack_list) != 0:
            print(stack_list[-1])
        else:
            print(-1)
```
