```python
n = int(input())
p_list = []
for _ in range(n):
    age, name = input().split()
    age = int(age)
    p_list.append([age, name])
p_list = sorted(p_list, key = lambda x : x[0])
for a, b in p_list:
    print(a, b)
```

```python
# 조금 더 짧은 코딩
n = int(input())
p_list = []
for _ in range(n):
    p_list.append(list(input().split()))
p_list = sorted(p_list, key = lambda x : int(x[0]))
for a, b in p_list:
    print(a, b)
```
