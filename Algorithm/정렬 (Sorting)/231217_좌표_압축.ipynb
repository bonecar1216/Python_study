```python
n = int(input())
num_list = list(map(int,input().split()))
num_set = set(num_list)
num_dict = {}
for i, num in enumerate(sorted(num_set)):
    num_dict[num] = i
for a in num_list:
    print(num_dict[a], end = ' ')
```
