```python
# 내 풀이
n = int(input())
num_dict = {}
for _ in range(n):
    num = int(input())
    if num not in num_dict.keys():
        num_dict[num] = 1
    else:
        num_dict[num] += 1
num_dict_sorted = dict(sorted(num_dict.items(), key=lambda x: x[0]))
for key, values in num_dict_sorted.items():
    for _ in range(values):
        print(key)
```

```python
# 다른 사람 풀이
import sys

n = int(sys.stdin.readline())
num_list = [0] * 10001

for _ in range(n):
    num_list[int(sys.stdin.readline())] += 1

for i in range(10001):
    if num_list[i] != 0:
        for j in range(num_list[i]):
            print(i)
```

* 내 풀이는 딕셔너리를 활용해서 풀이를 했으나, 백준 홈페이지에 제출했을 때 시간 초과 발생
* 리스트를 활용하는 것보다 딕셔너리를 활용하는 것은 메모리를 줄이고 계산 시간을 절약할 수 있으나 딕셔너리를 정렬하는 것에서 시간 초과가 발생한 것으로 보임.
