```python
array = [7,5,9,0,3,1,6,2,4,8]

for i in range(1,len(array)):
    for j in range(i,0,-1): # i 인덱스 이전으로 가면서 숫자 비교
        if array[j] < array[j-1]: # 한 칸씩 왼쪽으로 이동
            array[j], array[j-1] = array[j-1], array[j]
        else: # 자기보다 작은 데이터를 만나면 그 위치에서 멈춤 (데이터가 이미 정렬이 되어있으므로, 추가적으로 비교할 필요 없음)
            break
print(array)
```

    * 삽입 정렬도 선택 정렬과 마찬가지로 시간 복잡도가 O(N**2)이지만, 현재 리스트가 거의 정렬이 되어있을 경우에는 매우 빠르게 동작함.
