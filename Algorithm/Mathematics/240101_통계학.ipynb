https://www.acmicpc.net/problem/2108

```python
import sys
n = int(sys.stdin.readline())
num = []
num_dict = {}
for _ in range(n):
    m = int(sys.stdin.readline())
    num.append(m)
    if m not in num_dict.keys():
        num_dict[m] = 1
    else:
        num_dict[m] += 1

num.sort()
key_list = [k for k,v in num_dict.items() if v == max(num_dict.values())]
key_list.sort()
print(round(sum(num)/len(num)))
print(num[int(len(num)/2)])
print(key_list[0] if len(key_list) == 1 else key_list[1])
print(num[-1]-num[0])
```
