https://www.acmicpc.net/problem/4948

```python
def find_prime(n):
    num_list = [i for i in range(n+1)]
    num_list[0], num_list[1] = 0, 0
    prime = []
    for i in range(2,n+1):
        if num_list[i] != 0:
            prime.append(i)
            for j in range(2*i, n+1, i):
                if num_list[j]%num_list[i]==0:
                    num_list[j] = 0
    return prime

while True:
    n = int(input())
    if n == 0:
        break
    elif n == 1:
        print(1)
    else:
        print(len(find_prime(2*n))-len(find_prime(n)))
```

* 위 코드는 매번 함수를 통해 계산해야 하므로 시간을 줄일 수 있는 방법으로 접근해보기로 함.
* 함수에서 계산하는 과정없이 소수의 갯수를 미리 계산해서 저장하고 있는 방식을 사용해보기로 함.

```python
max_num = 123456*2
num_list = [i for i in range(max_num+1)]
num_list[0], num_list[1] = 0, 0

for i in range(2,max_num+1):
    if num_list[i] != 0:
        for j in range(2*i, max_num+1, i):
            if num_list[j]%num_list[i]==0:
                num_list[j] = 0
        num_list[i] = 1

while True:
    n = int(input())
    if n == 0:
        break
    elif n == 1:
        print(1)
    else:
        print(sum(num_list[:2*n+1])-sum(num_list[:n+1]))
```

* 첫번재 코드는 백준 제출시 실행 시간이 4620ms인 반면, 두번째 코드는 388ms으로 나옴.
* 함수 계산 시간을 줄임으로써 실행시간이 10배이상 빨라짐.
