https://www.acmicpc.net/problem/17103

```python
prime = [False, False] + [True] * (1000001-2)

for i in range(2, 1000001):
    if prime[i]:
        for j in range(2*i, 1000001, i):
            prime[j] = False

t = int(input())

for _ in range(t):
    num = int(input())
    cnt = 0
    for j in range(2,num//2+1):
        if prime[j] and prime[num-j]:
            cnt += 1
    print(cnt)
```
