```python
def gcd(a,b):
    if a%b == 0:
        return b
    elif b == 0:
        return a
    else:
        return gcd(b,a%b)

a, b = map(int,input().split())
c, d = map(int,input().split())

denominator = b*d//gcd(b,d)
numerator = a * denominator//b + c * denominator//d

print(numerator//gcd(numerator, denominator), denominator//gcd(numerator, denominator))
```
