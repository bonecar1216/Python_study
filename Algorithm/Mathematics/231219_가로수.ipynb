```python
def gcd(a,b):
    if a%b == 0:
        return b
    elif b == 0:
        return a
    else:
        return gcd(b,a%b)

n = int(input())
tree_pos = []
dist = []
for i in range(n):
    tree_pos.append(int(input()))
    if i != 0:
        dist.append(tree_pos[i]-tree_pos[i-1])
dist_set = sorted(set(dist))

r = dist_set[0]
for i in range(1,len(dist_set)):
    r = gcd(r,dist_set[i])

res = 0
for j in range(len(dist)):
    res += (dist[j]//r - 1)
print(res)
```
