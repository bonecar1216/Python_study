스크린샷 2024-01-07 오전 12.45.48.png

https://www.acmicpc.net/problem/1932

```python
n = int(input())
dp = []
for _ in range(n):
    dp.append(list(map(int,input().split())))

for i in range(1,n):
    for j in range(i+1):
        if j == 0:
            dp[i][j] += dp[i-1][j]
        elif i==j:
            dp[i][j] += dp[i-1][j-1]
        else:
            dp[i][j] += max(dp[i-1][j-1], dp[i-1][j])

print(max(dp[n-1]))
```
