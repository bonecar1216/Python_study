```python
import sys

def fib(n):
    global a
    if n == 1 or n == 2:
        a += 1
        return 1
    else:
        return fib(n-1)+fib(n-2)

def fibonacci(n):
    global b
    f = [0]*41
    f[1] = 1
    f[2] = 1
    for i in range(3,n+1):
        f[i] = f[i-1]+f[i-2]
        b += 1
    return f[i]


n = int(sys.stdin.readline())
a,b = 0,0
fib(n)
fibonacci(n)
print(a, b)
```
