```python
word = []
for _ in range(5):
    a = input()
    a += (15-len(a)) * ' '
    word.append(a)

res = ''
for j in range(15):
    for i in range(5):
        if word[i][j] != ' ':
            res += word[i][j]

res.replace(' ','')
print(res)
```

```python
# 다른 사람 코딩
# 출처: https://velog.io/@notnavi/%EB%B0%B1%EC%A4%80-10798-%EC%84%B8%EB%A1%9C%EC%9D%BD%EA%B8%B0
words = [input() for i in range(5)]

for j in range(15):
    for i in range(5):
        if j < len(words[i]):
            print(words[i][j], end='')

# 문자열이 끝나면 print가 진행되지 않도록 함. 내가 코딩한 것보다 짧은 코딩으로 구현하였으며, 불필요한 연산 과정이 없음.
```
