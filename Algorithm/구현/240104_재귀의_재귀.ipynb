```python
import sys
input = sys.stdin.readline()

def recursion(s, l, r):
    if l >= r: return 1,l
    elif s[l] != s[r]: return 0,l
    else: return recursion(s, l+1, r-1)

def isPalindrome(s):
    return recursion(s, 0, len(s)-1)

t = int(input())
for _ in range(t):
    word = input()
    n, m = isPalindrome(word)
    print(n, m+1)

```
