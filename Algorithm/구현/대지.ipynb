스크린샷 2023-12-13 오후 10.55.55.png

```python
n = int(input())

x_min = 10001
y_min = 10001
x_max = -10001
y_max = -10001

if n < 2:
    x, y = map(int, input().split())
    print(0)
else:
    for _ in range(n):
        x, y = map(int, input().split())
        if x > x_max:
            x_max = x
        if x < x_min:
            x_min = x
        if y > y_max:
            y_max = y
        if y < y_min:
            y_min = y

    print((x_max-x_min)*(y_max-y_min))
```
