https://www.acmicpc.net/problem/3009

```python
a_dict = {}
b_dict = {}

for _ in range(3):
    a, b = map(int,input().split())
    if a not in a_dict:
        a_dict[a] = 1
    else:
        a_dict[a] += 1

    if b not in b_dict:
        b_dict[b] = 1
    else:
        b_dict[b] += 1

a_dict = {v:k for k,v in a_dict.items()}
b_dict = {v:k for k,v in b_dict.items()}

print(a_dict[1], b_dict[1])
```
