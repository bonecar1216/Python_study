```python
import sys

# n=int(sys.stdin.readline())
n=int(input())
chat_list = []
ans = 0
for _ in range(n):
    # chat = sys.stdin.readline()
    chat = input()
    if chat == 'ENTER':
        ans += len(set(chat_list))
        chat_list = []
    else:
        chat_list.append(chat)

ans += len(set(chat_list))
print(ans)
```
