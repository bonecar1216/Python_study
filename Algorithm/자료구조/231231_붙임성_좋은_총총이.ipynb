```python
import sys

n = int(sys.stdin.readline())
dance_member = ['ChongChong']
for _ in range(n):
    a,b = sys.stdin.readline().split()
    if a in dance_member:
        dance_member.append(b)
    elif b in dance_member:
        dance_member.append(a)
print(len(set(dance_member)))
```
