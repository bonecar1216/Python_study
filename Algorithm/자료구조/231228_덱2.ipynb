```python
import sys
from collections import deque

n = int(sys.stdin.readline())
# n = int(input())
queue = deque([])
for _ in range(n):
    command = sys.stdin.readline().split()
    # command = input().split()
    if command[0] == '1':
        queue.appendleft(int(command[1]))
    elif command[0] == '2':
        queue.append(int(command[1]))
    elif command[0] == '3' and len(queue) != 0:
        print(queue.popleft())
    elif command[0] == '4'and len(queue) != 0:
        print(queue.pop())
    elif command[0] == '5':
        print(len(queue))
    elif command[0] == '6':
        if len(queue) == 0:
            print(1)
        else:
            print(0)
    elif command[0] == '7' and len(queue) != 0:
        print(queue[0])
    elif command[0] == '8' and len(queue) != 0:
        print(queue[-1])
    else:
        print(-1)
```
