스크린샷 2023-12-27 오후 9.39.51.png

https://www.acmicpc.net/problem/18258

```python
n = int(input())
queue = []
for _ in range(n):
    command = input().split()
    if command[0] == 'push':
        queue.append(command[1])
    elif command[0] == 'pop' or command[0] == 'front' or command[0] == 'back':
        if len(queue) == 0:
            print(-1)
        else:
            if command[0] == 'pop':
                print(queue.pop(0))
            elif command[0] == 'front':
                print(queue[0])
            else:
                print(queue[-1])
    elif command[0] == 'size':
        print(len(queue))
    elif command[0] == 'empty':
        if len(queue) == 0:
            print(1)
        else:
            print(0)
```

* 리스트를 이용하여 큐를 구현하려고 하였으나, 리스트에서 값이 빠질 때마다 재정렬하는 과정에서 연산이 많아져 시간 초과가 발생함.
* deque를 이용하여 문제를 풀어보기로 함.

```python
from collections import deque
import sys

n = int(sys.stdin.readline())
queue = deque([])
for _ in range(n):
    command = sys.stdin.readline().split()
    if command[0] == 'push':
        queue.append(command[1])
    elif command[0] == 'pop' or command[0] == 'front' or command[0] == 'back':
        if len(queue) == 0:
            print(-1)
        else:
            if command[0] == 'pop':
                print(queue.popleft())
            elif command[0] == 'front':
                print(queue[0])
            else:
                print(queue[-1])
    elif command[0] == 'size':
        print(len(queue))
    elif command[0] == 'empty':
        if len(queue) == 0:
            print(1)
        else:
            print(0)
```
